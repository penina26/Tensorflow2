```python
import pandas as pd
import numpy as np

import seaborn as sns

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import load_model
```

```python
df = pd.read_csv('DATA/fake_reg.csv')
```

```python
df.head()
```

```python
#create a pairpolot of the dataframe
sns.pairplot(df);
```

```python
from sklearn.model_selection import train_test_split
```

```python
#tensoflow works with numpy arrays so we need to convert the seeries/dataframe to array by adding .values 
X = df[['feature1', 'feature2']].values
y = df['price'].values
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape, X_test.shape)
```

```python
#Normalizing data
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
#transform the x_test
X_test = scaler.transform(X_test)
```

```python
#create a neural network
model = Sequential()

model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(4, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='rmsprop', loss='mse')
```

```python
model.fit(x=X_train, y=y_train, epochs=250)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
model.evaluate(X_train, y_train, verbose=0)
```

```python
test_predictions = model.predict(X_test)
```

```python
test_predictions
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
```

```python
pred_df = pd.DataFrame(y_test, columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df, test_predictions], axis=1)
```

```python
pred_df.columns = ['Test True Y', 'Model Predictions']
```

```python
pred_df
```

```python
sns.scatterplot(x='Test True Y',y='Model Predictions', data=pred_df);
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
 mean_absolute_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```

```python
mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions'], squared=False)
```
